```python
#import excel document to python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
  
df = pd.read_excel("kpmg_new_customers_list.xlsx")
print(df)
```

```python
#delete all columns with null values

df=df.dropna(axis=1,how='all')
```

```python
#delete columns without a column name

df = df.drop(['Column1','Column2','Column3','Column4','Column5'], axis=1)
df
```
